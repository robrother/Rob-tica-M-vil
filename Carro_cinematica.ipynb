```python
# importar el módulo pyplot
import matplotlib.pyplot as plt
import math
```

```python
#Puntos iniciales
x = [1, -1, -1, 1]   
y = [0, -1, 1, 0]   

#Puntos iniciales con la rotacion teta
x0=[]
y0=[]

#Puntos futuros con o sin la rotacion w
x1=[]
y1=[]

#condiciones iniciales
teta = 45 # angulo inicial
delta = 1  # instante de tiempo
v = 1     # velociad del carro   
w = 0 #ángulo de giro  


if (w == 0):
    print("omega igual a cero, el vehiculo no rota")
    for element in zip (x,y):
        xrot = element[0]*math.cos(teta) - element[1]*math.sin(teta)
        yrot = element[0]*math.sin(teta) + element[1]*math.cos(teta)
        x0.append(xrot)
        y0.append(yrot)
        
        #xfutu = element[0] + delta*(v*math.cos(teta))
        #yfutu = element[1] + delta*(v*math.sin(teta))
        xfutu = xrot + delta*(v*math.cos(teta))
        yfutu = yrot + delta*(v*math.sin(teta))
        x1.append(xfutu)
        y1.append(yfutu)
    
    print(x0)
    print(y0)
    
elif (w != 0):
    print("omega diferente de cero, el vehículo presenta una rotación")
    for element in zip (x,y):
        xrot = element[0]*math.cos(teta) - element[1]*math.sin(teta)
        yrot = element[0]*math.sin(teta) + element[1]*math.cos(teta)
        x0.append(xrot)
        y0.append(yrot)
        
        #xfutu = element[0] + delta*(v*math.cos(teta))
        #yfutu = element[1] + delta*(v*math.sin(teta))
        xfutu = xrot + delta*(v*math.cos(teta))
        yfutu = yrot + delta*(v*math.sin(teta))
        xrot1 = xfutu*math.cos(w) - yfutu*math.sin(w)   #
        yrot1 = xfutu*math.sin(w) + yfutu*math.cos(w)   #
        x1.append(xrot1) #xfutu
        y1.append(yrot1) #yfutu
    
    print(x1)
    print(y1)

    
plt.plot(x,y, marker="o")   # Dibuja el gráfico
plt.plot(x1,y1, marker="o")   # Dibuja el gráfico
plt.plot(x0,y0, marker="o")   # Dibuja el gráfico

plt.title("Gráfica de un solo punto")   # Establece el título del gráfico
plt.xlabel("X")   # Establece el título del eje x 
plt.ylabel("Y")   # Establece el título del eje y-1
```
