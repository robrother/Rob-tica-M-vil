```python
# importar el módulo pyplot
import matplotlib.pyplot as plt
```

```python
x = [11]   
y = [10]   
plt.plot(x,y, marker="o")   # Dibuja el gráfico
plt.title("Gráfica de un solo punto")   # Establece el título del gráfico
plt.xlabel("abscisa")   # Establece el título del eje x 
plt.ylabel("ordenada")   # Establece el título del eje y
```
