```python
# importar el m√≥dulo pyplot
import matplotlib.pyplot as plt
```

```python
x = [11]   
y = [10]   
plt.plot(x,y, marker="o")   # Dibuja el gr√°fico
plt.title("Gr√°fica de un solo punto")   # Establece el t√≠tulo del gr√°fico
plt.xlabel("abscisa")   # Establece el t√≠tulo del eje x 
plt.ylabel("ordenada")   # Establece el t√≠tulo del eje y
```
